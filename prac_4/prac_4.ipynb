```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , mean_squared_error
import numpy as np
```

```python
df=pd.read_csv('boston-housing.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.columns
```

```python
x=df[['ID','crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat']]
```

```python
y=df['medv']
```

```python
x_train, x_test, y_train, y_test,= train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
r2_Score = r2_score(y_test,y_pred)
r2_Score
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
```

```python
coef=np.polyfit(y_test,y_pred,1)
poly= np.poly1d(coef)
y_fit=poly(y_test)
```

```python
plt.scatter(x=y_test,y=y_test)
plt.plot(y_test,y_fit,color='red')
plt.xlabel('true values')
plt.ylabel('predictation')
plt.show()
```
