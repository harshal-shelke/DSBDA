```python
top_10=df.nlargest(15,'2022 Population')
```

```python
print(top_10)
```

```python
plt.figure(figsize=(12, 5))
plt.bar(top_10['Country/Territory'], top_10['2022 Population'])
plt.xlabel('Country/Territory')
plt.ylabel('Population')
plt.title('Top 10 Countries by 2022 Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
